```python
def plotfile(name, k,loop,guard,parameter):
    filename = f"./{name}_{k}_{loop}_{guard}.csv"
    df = pd.read_csv(filename, index_col=None, header=0)
    
    label = f'k:{k} loop:{loop} guard:{guard}' 
    plt.plot('N', parameter, data=df.loc[ df['Heuristics'] == True,], linestyle='dashed', label=f"True {label}")
    plt.plot('N', parameter, data=df.loc[ df['Heuristics'] == False,], label=f'False {label}')
```

```python
def plotTotalTime(name,k, loop,guard):
    plotfile(name,k,loop,guard, 'Total time')
    
def plotVerification(name,k,loop,guard):
    plotfile(name,k,loop,guard, 'Time spent on verification')
```

```python
plotTotalTime('pullUp',50,10,10)
plotTotalTime('pullUp',50,10,5)

plotTotalTime('pullUp',50,10,15)
plotTotalTime('pullUp',50,20,10)
plotTotalTime('pullUp',50,20,5)

plotTotalTime('pullUp',50,20,15)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
